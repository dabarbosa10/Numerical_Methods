Using three different numerical methods, the differential 
equation 
$$\frac{dy}{dx}=2x^2-4x+y$$
will be resolved.

```python
def funcion(x,y):
    r=2*x**2-4*x+y
    return r
```

```python
def euler(n,xf,x0,y0):
    h=(xf-x0)/n
    Y=[]
    X=[]
    X.append(x0)
    Y.append(y0)
    for i in np.linspace(x0+h,xf,n):
        yi=y0+h*funcion(x0,y0)
        x0=i
        y0=yi
        X.append(x0)
        Y.append(y0)
    Xv=np.array(X)
    Yv=np.array(Y)
    return[Xv,Yv]
```

```python
X,Y=euler(20,3.0,1.0,0.7182)
```

We know the analytical solution of the above differential equation, the 
solution is 
$$y=e^x-2x^2$$

```python
def analitica(x):
    y=np.exp(x)-2*x**2
    return y
```

```python
x=np.linspace(1,3,21)
```

```python
y=analitica(x)
```

```python
plt.plot(X,Y,label='Euler')
plt.plot(x,y,label='Analytic')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Euler Method')
plt.legend()
plt.show
```

```python
def heun(x0,xf,y0,n):
    h=(xf-x0)/n
    X=[]
    Y=[]
    X.append(x0)
    Y.append(y0)
    xi=x0+h
    for i in np.linspace(xi,xf,n):
        K1=h*funcion(x0,y0)
        K2=h*funcion(x0+h,y0+K1)
        yi=y0+0.5*(K1+K2)
        x0=i
        y0=yi
        X.append(x0)
        Y.append(y0)
    Xv=np.array(X)
    Yv=np.array(Y)
    return[Xv,Yv]
        
```

```python
X1,Y1=heun(1.0,3.0,0.7182,20)
```

```python
plt.plot(X1,Y1,'kD',label='Heun')
plt.plot(X,Y,label='Euler')
plt.plot(x,y,label='Analytic')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Different Methods')
plt.legend()
plt.show()
```
